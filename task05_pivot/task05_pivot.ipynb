```python
import pandas as pd
import numpy as np

#  Đọc dữ liệu hoàn chỉnh từ Task 04
df = pd.read_csv('../data_clean/full_dataset.csv')
# 1. Tổng lượng điện tiêu thụ theo house_type × tariff_level
pivot_consumption = pd.pivot_table(
    df,
    values='consumption_kwh',
    index='house_type',
    columns='tariff_level',
    aggfunc='sum'
)

print("\n--- Pivot: Tổng điện tiêu thụ theo Loại nhà × Bậc thang ---")
print(pivot_consumption)

#  2.Tiền điện trung bình theo region × house_type
pivot_cost = pd.pivot_table(
    df,
    values='price_per_kwh',
    index='region',
    columns='house_type',
    aggfunc='mean'
)

print("\n--- Pivot: Tiền điện trung bình theo Khu vực × Loại nhà ---")
print(pivot_cost)

# 3. Stack / Unstack để hiểu MultiIndex
print("\n--- Stack pivot_consumption ---")
stacked = pivot_consumption.stack()
print(stacked)

print("\n--- Unstack lại pivot_consumption ---")
unstacked = stacked.unstack()
print(unstacked)

# 4. Phân tích nhóm tiêu thụ nhiều điện nhất
# House_type tiêu thụ nhiều điện nhất (tổng)
house_type_max = (
    pivot_consumption.sum(axis=1)
    .idxmax()
)

# Bậc thang có tổng tiêu thụ cao nhất
tariff_max = (
    pivot_consumption.sum(axis=0)
    .idxmax()
)

print("\n--- Nhận xét ---")
print(f"Loại nhà tiêu thụ nhiều điện nhất: {house_type_max}")
print(f"Bậc thang giá có lượng điện tiêu thụ cao nhất: {tariff_max}")
```
