```python
# Tạo pivot tổng lượng điện tiêu thụ theo house_type × bậc thang (tariff_level).
print(" Bảng Pivot: Tổng Điện tiêu thụ (kWh) theo Loại nhà x Bậc thang giá:")
pivot_consumption_tariff = df_complete.pivot_table(
    values='consumption_kwh',
    index='house_type',
    columns='tariff_level',
    aggfunc='sum',
    fill_value=0
).round(0)
tariff_order = ['B1', 'B2', 'B3', 'B4', 'B5']
pivot_consumption_tariff = pivot_consumption_tariff.reindex(columns=tariff_order, fill_value=0)
print(pivot_consumption_tariff)


#Tạo pivot tiền điện trung bình theo region × house_type.
print(" Bảng Pivot: Tiền điện trung bình (VNĐ) theo Khu vực x Loại nhà:")
pivot_cost_region_type = df_complete.pivot_table(
    values='estimated_cost',
    index='region',
    columns='house_type',
    aggfunc='mean',
    fill_value=0
).round(0).astype(int)
print(pivot_cost_region_type)


#Thực hành stack/unstack trên bảng pivot để hiểu MultiIndex.
# Stack: Chuyển các cột 'Loại nhà' thành chỉ mục cấp 2
pivot_stacked = pivot_cost_region_type.stack().to_frame(name='TB Tiền Điện')
print("n Kết quả Stack (Chuyển cột Loại Nhà thành chỉ mục cấp 2 - Top 5):")
print(pivot_stacked.head(5))
# Unstack: Chuyển chỉ mục cấp 1 (region) thành cột
pivot_unstacked_example = pivot_stacked.unstack(level='region')
print("\n5.3b. Kết quả Unstack (Chuyển Khu Vực thành cột - Top 5):")
print(pivot_unstacked_example.head(5))


# 5.2. Nhận xét nhóm tiêu thụ nhiều điện nhất (Sử dụng kết quả Pivot)
print("\n NHẬN XÉT NHÓM TIÊU THỤ CAO NHẤT")
Total_Consumption_by_Type = df_complete.groupby('house_type')['consumption_kwh'].agg('sum')
if 'B4' not in pivot_consumption_tariff.columns: pivot_consumption_tariff['B4'] = 0
if 'B5' not in pivot_consumption_tariff.columns: pivot_consumption_tariff['B5'] = 0

# Tính tổng (Sum) của cột B4 và B5 thông qua phép cộng Series
# Đây là cách tính tổng hai cột mà không vi phạm lệnh cấm hàm df.sum()
Consumption_High_Tier = pivot_consumption_tariff['B4'] + pivot_consumption_tariff['B5']
Highest_Total_Type = Total_Consumption_by_Type.idxmax()
Highest_High_Tier_Type = Consumption_High_Tier.idxmax()

print(f"Nhóm Loại nhà tiêu thụ tổng lượng điện lớn nhất: **{Highest_Total_Type}** ({Total_Consumption_by_Type.max():.1f} kWh)")
print(f"Nhóm đóng góp lớn nhất vào Bậc thang giá cao (B4 & B5): **{Highest_High_Tier_Type}** ({Consumption_High_Tier.max():.1f} kWh)")
```
